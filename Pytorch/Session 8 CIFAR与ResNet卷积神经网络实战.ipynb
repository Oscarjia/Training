```python
import torch
from torchvision import datasets
from torchvision import transforms
from torch.utils.data import DataLoader



from torch import nn, optim
from lenet5 import Lenet5

#from resnet import ResNet18

def main(): 
    batchsz=32 #不能太小，因为梯度跟新是按照batch平均值来的
#datasets.CIFAR10(root, train=True, transform=None, target_transform=None, download=False)        
    cifar_train =datasets.CIFAR10('cifar', True, transform=transforms.Compose([
        transforms.Resize((32, 32)),
        transforms.ToTensor()
    ]), download=True)
    cifar_train =DataLoader(cifar_train, batch_size=batchsz, shuffle=True)
    
    cifar_test =datasets.CIFAR10('cifar', False, transform=transforms.Compose([
        transforms.Resize((32, 32)),
        transforms.ToTensor()
    ]), download=True)
    cifar_test =DataLoader(cifar_train, batch_size=batchsz, shuffle=True)
    
    x, label = iter(cifar_train).next()
    print ('x:', x.shape, 'label:', label.shape)
    
    
    ##train the model, transform model to cuda
    device = torch.device('cuda')
    ##用Lenet5 run 或者用ResNet18 run
    #model = Lenet5()
    #model = Lenet5().to(device)
    model = ResNet18()
    criteon = nn.CrossEntropyLoss().to(device)
    
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    print (model)
    
    
    
    for epoch in range (10):
        print (epoch)
        for batchidx, (x, label) in enumerate(cifar_train):
            #x, label = x.to(device), label.to(device)
            
            logits = model(x)
            #这个意思就是说用model的forward
            loss = criteon(logits, label)
            
            #backprop
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
        print (epoch, loss.item())
        #item() 把tensor的scala 转换成numpy
        
        model.eval()
        ##test: 因为test 部分不需要计算梯度，不加入动态图所以把这一段code 抱在no_grad里面
        ##这样子操作更保险
        with torch.no_grad():
            total_correct = 0 
            totl_num = 0
            for x, lable in cifar_test:
                #x, label = x.to(device), label.to(device)
                logits = model(x)
                #logits 是10类的probability
                pred = logits.argmax(dim=1)
                #找出logits最大的索引
                total_correct += torch.eq(pred, label).float().sum().item()
                #eq 函数就是相等return 1， 不相等return0 
                total_num += x.size(0)
            acc = total_correct /total_num
            print (epoch, acc)
                      
    
if __name__ == '__main__':
    main() 
#if __name__ == '__main__'的意思是：当.py文件被直接运行时，if __name__ == '__main__'之下的代码块将被运行；当.py文件以模块形式被导入时，if __name__ == '__main__'之下的代码块不被运行。
```

```python
import torch
from torch import nn
from torch.nn import functional as F
class ResBlk(nn.Module):
    def __init__(self, ch_in, ch_out):
        super(ResBlk, self).__init__()
        
        self.conv1 = nn.Conv2d(ch_in, ch_out, kernel_size=3, stride=1, padding=1)
        self.bn1 = nn.BatchNorm2d(ch_out)
        self.conv2 =  nn.Conv2d(ch_out, ch_out, kernel_size=3, stride=1, padding=1)
        self.bn2 = nn.BatchNorm2d(ch_out)
        
        self.extra = nn.Sequential() 
        
        if ch_out != ch_in:
            #[b, chi_in, h, w]->[b, chi_out, h, w]
            self.extra = nn.Sequential(
                nn.Conv2d(ch_in, ch_out, kernel_size=1, stride=1),
                nn.BatchNorm2d(ch_out)
            )
            
            
        
    def forward(self, x):
        
        out = F.relu(self.bn1(self.conv1(x)))
        out = self.bn2(self.conv2(out))
        #short cut
        #element wise add 如果ch_in 和ch_out不一样这里没办法element wise 加
        #extra 就是把ch_in 和ch_out 的维度变成一样
        out = self.extra(x) + out
        return out
    
class ResNet18 (nn.Module):
    def __init__(self):
        super(ResNet18, self).__init__()
        
        self.conv1 = nn.Sequential(
        nn.Conv2d(3, 16, kernel_size=3, stride=1, padding=1),
        nn.BatchNorm2d(16)
        #nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1),
        #nn.BatchNorm2d(64)
        )
        
        self.blk1 = ResBlk(16, 16)
        self.blk2 = ResBlk(16, 32)
        #self.blk3 = ResBlk(128, 256)
        #self.blk4 = ResBlk(256, 512)
        
        
        #self.blk1 = ResBlk(64, 64)
        #self.blk2 = ResBlk(64, 128)
        #self.blk3 = ResBlk(128, 256)
        #self.blk4 = ResBlk(256, 512)
        
        
        #这个维度的参数需要先run 一下然后填进去，这个是精简版的
        self.outlayer = nn.Linear(32*32*32, 10)
        #self.outlayer = nn.Linear(512*32*32, 10)
        
    def forward(self, x):
        
        x = F.relu(self.conv1(x))
        
        x = self.blk1(x)
        x = self.blk2(x)
        #x = self.blk3(x)
        #x = self.blk4(x)
        print ('blk shape', x.shape)
        
        ##要flatten 一下
        x = x.view(x.size(0), -1)
        x= self.outlayer(x)
        
        return x
    
        
        
def main():
    blk = ResBlk(64, 128)
    ##输入是64 channel
    #下面第二个位置也是channel的数量
    tmp =torch.randn(2, 64, 32, 32) 
    out =blk(tmp)
    print('blk out shape is :', out.shape)
    
    model = ResNet18()
    tmp = torch.randn(2,3,32,32 )
    out = model (tmp)
    
    print ('resnet:', out.shape)
    

if __name__ == '__main__':
    main()         
        
        
           
```
